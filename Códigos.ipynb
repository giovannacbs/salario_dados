```python
import pandas as pd
```

```python
df = pd.read_csv("/content/pnad4t23limpo3.csv")
```

```python
df['ocup'] = round(df['ocup'],0).astype(str)
```

```python
df.head()
```

```python
def media_salario(age,region,race,education,ocupation,hours_work):
  med_h = df.query(f"sexo == 'Homem' & faixa_etaria == '{age}' & regiao == '{region}' & raca == '{race}' & educ == '{education}' & ocup == '{ocupation}' & hrs_trab == '{hours_work}'")['salario'].mean()
  med_m = df.query(f"sexo == 'Mulher' & faixa_etaria == '{age}' & regiao == '{region}' & raca == '{race}' & educ == '{education}' & ocup == '{ocupation}' & hrs_trab == '{hours_work}'")['salario'].mean()
  dif = round(((med_h/med_m)-1)*100,2)
  return round(med_h,1), round(med_m,1), dif
```

```python
media_salario("26-35","Norte", "Branca", "Ensino Médio", "71", "Full-time")
```
